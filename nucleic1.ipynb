## Output a QUAL file

The QUAL file format has header lines just like FASTA, followed by numerical quality scores separated by spaces.

FASTQ represents the quality score for each basepair using [ASCII](http://en.wikipedia.org/wiki/ASCII) characters. It is simple to convert these characters back to a numerical score, e.g. for the FASTQ quality score "h":


`````python
   print ord('h')-64
`````


The number 64 is the 'offset' used by Illumina (i.e. the number added to the score to produce the ASCII code).

We need to convert each qual string from the FASTQ file into a list of numbers. In Python it is easy to iterate over each character of a string using a FOR loop:


`````python
from cogent.parse.fastq import MinimalFastqParser

out_file = open("DNA_Sequence.qual", "w")

for label, seq, qual in MinimalFastqParser('Chan_Jonathan_single.fastq', strict=False):
    out_file.write('>' + label + '\n')   # add a line here to print the FASTA-style header line
    for x in qual:
        num = ord(x)-64
        out_file.write(str(num) + ' ') # add a line within this loop to print the numerical quality score as a number, followed by a space but no carriage return.
    out_file.write('\n\n')# this command just prints a carriage return after all of the scores.

out_file.close()
`````

## Sequence assembly with Phrap

Once you have extracted the sequences and quality scores into separate files, you can assemble your reads using the linux command
````
phrap XXXX.fna
````




Phrap will load the sequences and quality scores, assemble the reads and produce several output files with different extensions:

**`.contigs`** contains the contigs that phrap has managed to build from the sequence fragments.

**`.contigs.qual`** contains the quality scores assigned by phrap for each base in the sequence.

**`.singlets`** contains any sequences that phrap was unable to assemble into a contig.

**`.log`** is a log file of the assembly process.

**`.problems`** and **`.problems.qual`** contain data for any sequences for which there was a problem.



The only file you need is `.contigs`, as this contains the assembly itself. 

Change the extension to `.contigs.fasta` (use the **`mv`** command in linux) so that PyCogent will recognise this as a FASTA file.
